```python
def rearrange(lst1,lst2,lst3,lst4):
    lst = []
    for i in range(len(lst1)):
        lst.append([lst1[i],lst2[i],lst3[i],lst4[i]])
    return lst
```

```python
#Let lst1 be shapes, lst2 be color, lst3 be numbers, and lst4 be shading
#lst1 (shape): a--->oval, b--->squiggle, c--->diamond
#lst2 (color): a--->red, b--->green, c--->purple
#lst3 (numbers): a--->1, b--->2, c--->3
#lst4 (shading): a--->solid, b--->unfilled, c--->striped

def transform(lst):
    new_lst = []
    for card in lst:
        if card[3] == '0':
            shape = "diamond"
        if card[3] == '1':
            shape = "oval"
        if card[3] == "2":
            shape = "squiggle"
        if card[1] == '1':
            color = "red"
        if card[1] == '2':
            color = "green"
        if card[1] == "0":
            color = "purple"
        if card[0] == '0':
            number = "3"
        if card[0] == '1':
            number = "1"
        if card[0] == "2":
            number = "2"
        if card[2] == '1':
            shading = "solid"
        if card[2] == '2':
            shading = "unfilled"
        if card[2] == "0":
            shading = "striped"
        new_lst.append([shape, color, number, shading])
        """new_lst.append([output_latex([[shape, color, number, shading]]),''.join(lst[lst.index(card)]), shape, color, number, shading])"""
    return new_lst
```

```python
def translate(lst, type):
    new_lst = []
    if type == "shape":
        for item in lst:
            if item == 'a':
                shape = "oval"
            if item == 'b':
                shape = "squiggle"
            if item == "c":
                shape = "diamond"
            new_lst.append(shape)
    if type == "color":
        for item in lst:
            if item == 'a':
                color = "red"
            if item == 'b':
                color = "green"
            if item == "c":
                color = "purple"
            new_lst.append(color)
    if type == "number":
        for item in lst:
            if item == 'a':
                number = "1"
            if item == 'b':
                number = "2"
            if item == "c":
                number = "3"
            new_lst.append(number)
    if type == "shading":
        for item in lst:
            if item == 'a':
                shading = "oval"
            if item == 'b':
                shading = "squiggle"
            if item == "c":
                shading = "diamond"
            new_lst.append(shading)
    return new_lst
```
