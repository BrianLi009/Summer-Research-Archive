```python
# The Basel Problem: Calculate 1/1 + 1/4 + 1/9 + 1/16 + 1/25 + ...

import math;

sum = 0;
for n in range(1,100000):
    sum += 1/(n**2);
print("We get that the sume is: " + str(sum));
print("This gives us the approximation: " + str(math.sqrt(6*sum)));
```

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(-5,5,100)
y1 = np.sin(x)
y2 = np.cos(x)

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1) #vertical axis
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
#ax.spines['right'].set_color('none')
#ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
# plot the function
plt.plot(x,y1, 'r')
plt.plot(x,y2, 'r')

# show the plot
plt.show()

#reference: https://scriptverse.academy/tutorials/python-matplotlib-plot-function.html
```
